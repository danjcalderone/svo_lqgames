```python

### generating widgets....
vec_opts = ['axes','vecs','coords'];
shape_opts = ['x','xcoords','simplex','cube','sphere','ball1','ballinf'];
basis_opts1 = [0,1,2,3,4,5]
basis_opts2 = [6,7,8,9,10,11];
show_opts = ['Nash','SVO','basis']
svo_opts = ['svo1','svo2'];
scale_opts = ['unscaled','scaled'] 
show_menu = SelectMultiple(options=show_opts,value=show_opts[0:1],description='trajs:',disabled=False,rows=3,layout={'width':'110pt'})
svo_menu = Select(options=svo_opts,value=svo_opts[0],description='SVO values:',disabled=False,rows=3,layout={'width':'110pt'}) 
basis_menu1 = SelectMultiple(options=basis_opts1,value=basis_opts1[0:1],description='basis opts:',disabled=False,rows=6,layout={'width':'80pt'})
basis_menu2 = SelectMultiple(options=basis_opts2,value=[],description='',disabled=False,rows=6,layout={'width':'20pt'})
scale_buttons = RadioButtons(options=scale_opts,value=scale_opts[0],description='basis scaling:',disabled=False,layout={'width':'100pt'})
theta2_slider = FloatSlider(step=0.01,value=np.pi/8,min=0.0+0.001,max=np.pi/2 - 0.001,description='theta2',orientation='horizontal',layout={'width':'300pt'},continuous_update=False);


# loading function arguments from widgets...
ctrls = {};
ctrls['show_opts'] = show_menu
ctrls['svo_opts'] = svo_menu
ctrls['basis_opts1'] = basis_menu1
ctrls['basis_opts2'] = basis_menu2
ctrls['scale_opt'] = scale_buttons
ctrls['theta2_opt'] = theta2_slider;

# building dashboard
dash = HBox([show_menu,svo_menu,basis_menu1,basis_menu2,scale_buttons,theta2_slider]);
widgets = {**ctrls};
fins['widgets'] = widgets;
ctrls['ins'] = fixed(fins);

viz = interactive_output(frame_SVO,ctrls);
display(viz,dash);
```
